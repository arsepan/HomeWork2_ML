## Задание 4 (1 балла)

Найдите точки перегиба этих функций в Python.

```python
import sympy as sp

x = sp.Symbol('x')

functions = {
    'x**2 - 30*x + 1': x**2 - 30*x + 1,
    '-x**3 + 5*x - 7*x**2 + 17': -x**3 + 5*x - 7*x**2 + 17,
    'x**2 - 1/x': x**2 - 1/x,
    '1/x**5 + x**17 + 8': 1/x**5 + x**17 + 8,
}

for func_name, func in functions.items():
    print(f'\nТочки перегиба для {func_name}: {func}')

    first_derivative = sp.diff(func, x)

    second_derivative = sp.diff(first_derivative, x)
    print(f'  Вторая производная: {second_derivative}')

    inflection_points = sp.solve(sp.Eq(second_derivative, 0), x)
    print(f'  Критические точки (f''(x) = 0): {inflection_points}')

    third_derivative = sp.diff(second_derivative, x)
    for point in inflection_points:
        try:
            point_eval = point.evalf(chop=True)
            if point_eval.is_real:
                third_derivative_value = third_derivative.subs(x, point_eval).evalf()
                print(f"    x = {point_eval}: f'''(x) = {third_derivative_value}")
                if third_derivative_value != 0:
                    print('    Это точка перегиба')
                else:
                    print("    Точка не является точкой перегиба (f'''(x) = 0)")
            else:
                print(f'    Комплексное решение x = {point} пропускаем')
        except:
            pass
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, solve


def plot_extr_points(func, label, lambda_func, min_x, max_x):
    try:
        f_prime = diff(func, x)
        f_double_prime = diff(f_prime, x)

        extrema_x = solve(f_prime, x)
        extrema_y = [func.subs(x, val) for val in extrema_x]

        inflection_x = solve(f_double_prime, x)
        inflection_y = [func.subs(x, val) for val in inflection_x]

        f = lambda_func

        x_vals = np.linspace(min_x, max_x, max_x - min_x)
        y_vals = f(x_vals)

        plt.figure(figsize=(10, 6))

        plt.plot(x_vals, y_vals, label=label, color='blue')

        for x_extr, y_extr in zip(extrema_x, extrema_y):
            plt.scatter(float(x_extr), float(y_extr), color='red', label='Экстремум')
            plt.annotate(f'({x_extr:.2f}, {y_extr:.2f})', (float(x_extr), float(y_extr) + 5),
                        textcoords='offset points', xytext=(-20, 5), ha='center', color='red')

        for x_infl, y_infl in zip(inflection_x, inflection_y):
            plt.scatter(float(x_infl), float(y_infl), color='green', label='Точка перегиба')
            plt.annotate(f'({x_infl:.2f}, {y_infl:.2f})', (float(x_infl), float(y_infl) - 10),
                        textcoords='offset points', xytext=(-20, -15), ha='center', color='green')

        plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
        plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
        plt.xlabel('x', fontsize=12)
        plt.ylabel('f(x)', fontsize=12)
        plt.title('График функции с экстремумами и точками перегиба', fontsize=14)
        plt.legend(loc='upper right')
        plt.grid(True)

        plt.show()
    except:
        pass
```

## Задание 6 (1 балл)

Модифицируйте функцию, осуществляющую метод градиентного спуска с семинара: добавьте опцию останавливать метод, если абсолютное значение производной функции в точке < некоторого $\delta$ (delta добавьте в аргументы функции).

```python
def gradient_descent(x_start, learning_rate, epsilon, delta, num_iterations, lambda_func):
    x = sp.symbols('x')
    f = lambda_func
    df_x = sp.diff(f(x))

    x_curr = x_start
    trace = [x_curr]

    for i in range(num_iterations):
        grad = df_x.subs(x, x_curr)
        if abs(grad) < delta:
            print(f"Остановка: |f'(x)| = {abs(grad):.4f} < delta = {delta}")
            return x_curr, trace

        x_new = x_curr - learning_rate * grad
        trace.append(x_new)

        if abs(x_new - x_curr) < epsilon:
            print(f'Остановка: |x_new - x_curr| = {abs(x_new - x_curr):.4f} < epsilon = {epsilon}')
            return x_new, trace

        x_curr = x_new

    print('Достигнуто максимальное число итераций')
    return x_curr, trace
```

## Задание 8 (1.5 балла)

Реализуйте [метод Ньютона](https://en.wikipedia.org/wiki/Newton%27s_method_in_optimization) (метод второго порядка) для поиска минимума функции.

Про метод ещё можно почитать [здесь](https://math.semestr.ru/optim/minimum.php).

```python
def newton_method(f_expr, x_start, epsilon=1e-6, max_iterations=1000):
    x = sp.symbols('x')
    df = sp.diff(f_expr, x)
    ddf = sp.diff(df, x)

    x_curr = x_start
    trace = [x_curr]

    for i in range(max_iterations):
        grad = df.subs(x, x_curr)
        hess = ddf.subs(x, x_curr)

        if abs(grad) < epsilon:
            break

        if hess == 0:
            raise ValueError('Вторая производная равна нулю. Невозможно применить метод Ньютона')

        x_new = x_curr - grad / hess
        trace.append(x_new)

        if abs(x_new - x_curr) < epsilon:
            break

        x_curr = x_new

    x_min = x_curr
    f_min = f_expr.subs(x, x_min)
    return x_min, f_min, trace
```

```python
def gradient_descent(f_expr, x_start, learning_rate=0.1, epsilon=1e-6, max_iterations=1000):
    x = sp.symbols('x')
    df = sp.diff(f_expr, x)

    x_curr = x_start
    trace = [x_curr]

    for i in range(max_iterations):
        grad = df.subs(x, x_curr)

        if abs(grad) < epsilon:
            break

        x_new = x_curr - learning_rate * grad
        trace.append(x_new)

        if abs(x_new - x_curr) < epsilon:
            break

        x_curr = x_new

    print('Достигнуто максимальное число итераций')
    return x_curr, f_expr.subs(x, x_curr), trace, i + 1
```
